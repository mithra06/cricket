```python
import json

with open("py_challange_input.json", 'r') as json_file:
    json_load = json.load(json_file)

data = json_load['Cricket']
for ind in data:
    for i in ind['formats']:
        reg=i['regions']
    
    
        for match in reg:
            name=match["name"]
            for mat in match['matches']:
                match_id=mat['id']

                team1=mat["team1"]
                team2=mat["team2"]
                prov=mat["streaming"]["liveStream"]["provider"]

                result=mat['results']["winner"]
                http=mat["httpLink"]
                #print (match_id,name,team1,team2,prov,result,http)
                print(json.dumps({"match_id":match_id,
                                  "winner_of_the_match":result, 
                                  "team1":team1,
                                  "team2":team2,
                                  "livestream_provider":prov,
                                  "match_link":http,
                                  "region_name":name},indent=4))
            
        



```
